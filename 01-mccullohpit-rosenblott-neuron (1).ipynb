```python
from random import choice
from numpy import array, dot , random
import numpy as np
```

```python
#MCNEURON for AND
w = random.rand(2)
w[1] = 1
w[0] = 1
train_daa = [
    (array([0,0]), 0),
        (array([0,1]), 0),
    (array([1,0]), 0),
    (array([1,1]), 1),

]
```

```python
step_function = lambda x: 0 if x<2 else 1
```

```python
for x, _ in train_daa:
    result = dot(x,w)
    
    print("{}: {} -> {}".format(x[:2], result, step_function(result)))
```

```python
#Rosenblatt neuron

step_function = lambda x: 0 if x < 50 else 1

train_data = [
    (array([0,0,1]), 1),
        (array([0,1,1]), 0),
    (array([1,0,1]), 0),
    (array([1,1,1]), 1),

]

w= random.rand(3)

b = .1  #bias term

errors = []
eta = 0.1  #learning rate
n = 10000

for i in range(n):
    x, expected = choice(train_data)
    
    #w = np.append(w,b)
    
    result = dot(w, x)
    
    error = expected - step_function(result)
    errors.append(error)
    w += eta * error *x
    
    
for x, _ in train_data:
    result = dot(w, x)
    print("{}: {} -> {}".format(x[:3], result, step_function(result)))
    
```
